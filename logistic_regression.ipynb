```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
import numpy as np
from sklearn.svm import SVC
```

```python
dataset = pd.read_csv(root_folder+"sample-notebook/final_dataset.csv")
```

```python
Y = dataset['upgrade']
X = dataset.drop(['upgrade','line_id', 'sim_size', 'date_observed'], axis=1)
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size=0.2, shuffle=True, random_state = 2)
```

```python
log = LogisticRegression()
log.fit(Xtrain, Ytrain)
ypred = log.predict(Xtest)
acc = accuracy_score(Ytest, ypred)
f1 = f1_score(Ytest, ypred)
print(acc)
print(f1)
```

```python
## tunning for hyperparameters using GridSearchCV
params = {
    'penalty' : ['l2','ekasticnet', 'none'], 
    'solver' : ['newton-cg', 'lbfgs', 'sag', 'saga'], 
    'max_iter' : [100,300,500]
    }
grid = GridSearchCV(LogisticRegression(),param_grid=params, scoring='f1')
grid.fit(Xtrain, Ytrain)
```

```python
grid.best_params_
```

```python
ypred = grid.predict(Xtest)
f1 = f1_score(Ytest, ypred)
acc = accuracy_score(Ytest, ypred)
print(f"f1 score: {f1}\n accuracy: {acc}")
```

```python
confusion_matrix(Ytest, ypred, labels=[1,0])
```
